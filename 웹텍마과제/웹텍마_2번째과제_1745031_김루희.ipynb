```python
import requests
```

```python
# 1. 지난 주 실습 중 패스한 전자신문 사이트에서 “인공지능“ 검색 부분 완성

url = "https://search.etnews.com/etnews/search.php?kwd=인공지능"
params = { "kwd"  : "인공지능" }

res = requests.get(url, params=params)

print("응답코드", res.status_code)  # 에러여부 확인
```

```python
res.encoding  # 데이터 인코딩 방식 확인
```

```python
res.text   # 데이터 수집 내용 확인
```

```python
print(len(res.text))  # 데이터 수집 사이즈 확인
```

```python
#2. 디지털 타임즈(www.dt.co.kr) 사이트에서 “인공지능“ 검색 후 위와 동일하게 자료 수집

url = "http://search2.dt.co.kr/search.jsp"
params = { "txt"  : "인공지능" }

res = requests.get(url, params=params)

print("응답코드", res.status_code)  # 에러여부 확인
```
